# Working with real world data set using python and sql

Lets work to connect the data with database using sql


```python
import sqlite3
import pandas as pd
%load_ext sql
import numpy as np
```

```python
try :
 df=pd.read_csv('D:\\Data Science\\Juypter Notebooks\\ChicagoPublicSchools.csv')
 print('Successfully read csv file')
except Exception as e :
 print(f'Failed to load the file due to the following error {e}')
 
```

Converting the csv data to Sqlite database to perform queries using sql


```python
try :
 conn=sqlite3.connect('Chicago_public_school.db')
 cursor=conn.cursor()
 %sql sqlite:///Chicago_public_school.db
 print('Connection Successful')
except Exception as e :
  print(f'Failed to connect the database due to {e}')
```

```python
try :
 df.to_sql('chicago_public',conn,if_exists='replace',index=False)
 print('Successful added data to database with table name : Chicago_Public')
except Exception as e :
 print(f'Failed to  load the data due to the error {e}')
```

```python
import prettytable
prettytable.DEFAULT='DEFAULT'
```

```python
query='''select * from chicago_public limit 5'''
cursor.execute(query)
output=cursor.fetchall()
print(output)
%sql SELECT * from chicago_public limit 5
```

```python
query_1='''select count(*) from chicago_public where "Elementary, Middle, or High School"='ES'  '''
cursor.execute(query_1)
output_1=cursor.fetchall()
print(f'The number of elementary schools in chicago is {output_1[0][0]}')
```

```python
%sql SELECT max(SAFETY_SCORE) as Highest_Safety_Score from chicago_public
```

```python
query_3='''select * from chicago_public where safety_score=(select max(safety_score) from chicago_public)'''
cursor.execute(query_3)
output_3=cursor.fetchall()
for data in output_3:
    print(data)
```

```python
query_4='''select * from chicago_public where AVERAGE_STUDENT_ATTENDANCE=(select max(AVERAGE_STUDENT_ATTENDANCE) from chicago_public) '''
cursor.execute(query_4)
output_4=cursor.fetchall()
for item in output_4:
    print(item)
```
